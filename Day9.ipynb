```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

temp=[20,25,30,35,40]
sale=[13,21,25,35,38]

x=np.array([temp]).T
print(x)
print(x.shape)
```

```python
y=np.array(sale)
print(y)
print(y.shape)
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred=model.predict(x)
print(y_pred)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
new=model.predict([[90]])
print(new)
```

```python
mse=mean_squared_error(y,y_pred)
print(mse)
```

```python
plt.scatter(temp,sale,marker="*")
plt.plot(temp,y_pred,color='r')
plt.xlabel('temp')
plt.ylabel('sale')
plt.show()
```

```python
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(temp, sale)
temp = [20,25,30,35,40]
sale = [13,21,25,35,38]
print('slope', slope)
print('intercept', intercept)


#user defined fn
def myfunc(x):
  return slope * x + intercept

y_pred = list(map(myfunc,temp))

plt.scatter(temp,sale)
plt.plot(temp,y_pred)
plt.show()


newsale = myfunc(49)
print( newsale )
```
